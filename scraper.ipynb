```python
import subprocess
import json
import os
from pathlib import Path

def get_proxy_ips():
    # Try to get the base directory from __file__ (works when running as a script).
    # If __file__ is not defined (e.g. in a notebook), use the current working directory.
    try:
        base_dir = Path(__file__).parent.resolve()
    except NameError:
        base_dir = Path(os.getcwd()).resolve()

    print("Base directory:", base_dir)
    
    # Define the subfolder relative to the base directory.
    subfolder = base_dir / "SetupProxyCloud"
    print("Looking for SetupProxyCloud at:", subfolder)
    
    if not subfolder.exists():
        print(f"Subfolder not found: {subfolder}")
        return []
    
    original_dir = os.getcwd()
    try:
        os.chdir(str(subfolder))
        print("Changed directory to:", os.getcwd())
        
        # Run the Terraform output command and capture the JSON output.
        result = subprocess.run(
            ["terraform", "output", "-json"],
            capture_output=True,
            text=True,
            check=True
        )
        outputs = json.loads(result.stdout)
        proxy_ips = outputs.get("proxy_ips", {}).get("value", [])
        return proxy_ips
    except subprocess.CalledProcessError as e:
        print("Error executing terraform output:", e)
        return []
    except json.JSONDecodeError as e:
        print("Error decoding JSON output:", e)
        return []
    finally:
        os.chdir(original_dir)

if __name__ == "__main__":
    proxies = get_proxy_ips()
    if proxies:
        print("List of Proxy IPs:")
        for ip in proxies:
            print(ip)
    else:
        print("No proxy IPs found.")
```

```python
# Assume df_combined is your combined DataFrame with the following columns:
# Date, Open, High, Low, Close, Adj Close, Volume, Symbol

# Ensure the Date column is in datetime format
df_combined['Date'] = pd.to_datetime(df_combined['Date'])

# Sort the DataFrame by Symbol and Date so that pct_change is calculated correctly
df_combined = df_combined.sort_values(["Ticker", "Date"])

# Calculate the daily percentage change on the 'Close' price for each stock (grouped by Symbol)
df_combined['Return'] = df_combined.groupby("Ticker")['Close'].pct_change() * 100

# Drop rows where the return is NaN (typically the first row for each stock)
df_returns = df_combined.dropna(subset=['Return'])

# Create an interactive line plot with Plotly Express
fig = px.line(
    df_returns,
    x="Date",
    y="Return",
    color="Ticker",
    title="Daily % Change in Close Price by Stock",
    labels={"Return": "Daily Return (%)"}
)

fig.show()
```
